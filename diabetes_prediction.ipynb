# Diabetes Prediction using Machine Learning

This project predicts whether a person is diabetic or non-diabetic using Machine Learning techniques. The model is trained on the PIMA Indians Diabetes Dataset and uses Logistic Regression for prediction.

## 1. Importing Required Libraries

In this step, we import the necessary Python libraries for data handling, preprocessing, model training, and evaluation.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

## 2. Loading the Dataset

The diabetes dataset is loaded into a Pandas DataFrame for further analysis and preprocessing.

```python
diabetes_data = pd.read_csv('diabetes.csv')

diabetes_data.head()
```

## 3. Data Preprocessing

Some medical attributes contain invalid zero values. These values are replaced with the median to improve data quality and model performance.

```python
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in cols:
    diabetes_data[col] = diabetes_data[col].replace(
        0,
        diabetes_data[col].median()
    )

diabetes_data.head()
```

## 4. Splitting Features and Target Variable

The independent variables (features) and dependent variable (Outcome) are separated for model training.

```python
X = diabetes_data.drop(columns='Outcome', axis=1)
Y = diabetes_data['Outcome']
```

## 5. Data Standardization

Feature scaling is performed to normalize the data and improve model performance.

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

## 6. Splitting Training and Testing Data

The dataset is divided into training and testing sets to evaluate model performance effectively.

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    stratify=Y,
    random_state=10
)
```

## 7. Model Training

A Logistic Regression model is trained using the training dataset.

```python
model = LogisticRegression()

model.fit(X_train, Y_train);
```

## 8. Model Evaluation

The trained model is evaluated using accuracy score on both training and testing datasets.

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction, Y_train)

print("Training Accuracy:", round(training_accuracy*100,2),'%')

X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction, Y_test)

print("Testing Accuracy:", round(testing_accuracy*100,2),'%')
```

## 9. Building a Prediction System

The trained model predicts whether a person is diabetic or non-diabetic based on input medical values.

```python
input_data = pd.DataFrame(
    [[166, 72, 19, 175, 25.8, 0.587, 51]],
    columns=[
        'Glucose',
        'BloodPressure',
        'SkinThickness',
        'Insulin',
        'BMI',
        'DiabetesPedigreeFunction',
        'Age'
    ]
)

scaled_input = scaler.transform(input_data)

prediction = model.predict(scaled_input)

if prediction[0] == 0:
    print("The person is not diabetic")
else:
    print("The person is diabetic")
```
